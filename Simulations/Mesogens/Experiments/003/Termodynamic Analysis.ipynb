# Termodynamic properties of the system

```python
import numpy
from matplotlib import pyplot
```

```python
data = numpy.genfromtxt(fname = 'T_8.0_restarted_trayectory.log', skip_header=True);
```

```python
pyplot.figure(figsize=(4,2.2), dpi=140);
pyplot.plot(data[:,0], data[:,10]);
pyplot.xlabel('time step');
pyplot.ylabel('Pressure');
pyplot.suptitle('P = 1.0, P = 0.1')
```

```python
mean_p = numpy.mean(data[:,10])
```

```python
print(mean_p)
```

```python
std_p = numpy.std(data[:,10])
```

```python
print(std_p)
```

```python
pyplot.figure(figsize=(4,2.2), dpi=140);
pyplot.plot(data[:,0], data[:,9]);
pyplot.xlabel('Time step');
pyplot.ylabel('Temperature');
pyplot.suptitle('T = 7.5, P = 0.1')
```

```python
mean_temp = numpy.mean(data[:,9])
```

```python
std_temp = numpy.std(data[:,9])
```

```python
print(mean_temp, std_temp)
```
