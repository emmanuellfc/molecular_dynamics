# Termodynamic Analysis

Importa los paquetes necesarios.

```python
import numpy
from matplotlib import pyplot
```

Debido a que este notebook es utilizado tanto para las rampas, como para los equilibrios del sistema, es conveniente definir una variable en la que asignemos el nombre del archivo del cual vamos a extraer la informacion sobre la termodinámica.

La estructura del nombre del archivo es la siguiente T + valor de la presión + P + valor de la presión + parte de la simulación, los nombres disponibles para "parte de la simulación", son los siguientes : initialization, ramp & stabilization.

```python
t_max = 5.0;
p_max = 1.4;
simulation_part = "stabilization";
```

```python
file_name = "T_" + str(t_max) + "_P_" + str(p_max) + "_" + simulation_part + ".log"
```

## Values of Pressure & Temperature

```python
data = numpy.genfromtxt(fname = file_name, skip_header=True);
```

```python
mean_p = numpy.mean(data[:,10]);
std_p = numpy.std(data[:,10]);
mean_temp = numpy.mean(data[:,9])
std_temp = numpy.std(data[:,9])
```

```python
print("P =", mean_p, "±",std_p)
```

```python
print("T =", mean_temp, "±", std_temp)
```

# Graphics

```python
pyplot.figure(figsize = (5,3), dpi = 140);
pyplot.plot(data[:,0], data[:,10], "g");
pyplot.xlabel('Time step');
pyplot.ylabel('Pressure');
pyplot.suptitle(simulation_part + " - "+ "T = " + str(t_max) + ' & ' + "P = " + str(p_max));
pyplot.grid(True, color = 'k', linestyle = '--', linewidth = 0.1);
```

```python
pyplot.figure(figsize = (5,3), dpi = 140);
pyplot.plot(data[:,0], data[:,9], "r");
pyplot.xlabel('Time step');
pyplot.ylabel('Temperature');
pyplot.suptitle(simulation_part + " - "+ "T = " + str(t_max) + ' & ' + "P = " + str(p_max));
pyplot.grid(True, color = 'k', linestyle = '--', linewidth = 0.1);
```

```python
pyplot.figure(figsize = (5,3), dpi = 140);
pyplot.plot(data[:,0], data[:,5]);
pyplot.xlabel('Time step');
pyplot.ylabel('Potential energy');
pyplot.suptitle("Stabilization - T = " + str(t_max) + ' & ' + "P = " + str(p_max));
```

```python
pyplot.figure(figsize = (5,3), dpi = 140);
pyplot.plot(data[:,0], data[:,6]);
pyplot.xlabel('Time step');
pyplot.ylabel('Kinetic energy');
pyplot.suptitle(simulation_part + " - "+ "T = " + str(t_max) + ' & ' + "P = " + str(p_max));
```
