```python
import fresnel
import math
```

```python
scene = fresnel.Scene()
```

```python
geometry = fresnel.geometry.Sphere(scene, N=16, radius=1.0)
```

```python
geometry.position[:] = [[2,1,1],
                        [4,1,1],
                        [6,1,1],
                        [8,1,1],
                        [1,1 + math.sqrt(3),1],
                        [3,1 + math.sqrt(3),1],
                        [5,1 + math.sqrt(3),1],
                        [7,1 + math.sqrt(3),1],
                        [2,1 + 2 * math.sqrt(3),1],
                        [4,1 + 2 * math.sqrt(3),1],
                        [6,1 + 2 * math.sqrt(3),1],
                        [8,1 + 2 * math.sqrt(3),1],
                        [1,1 + 3 * math.sqrt(3),1],
                        [3,1 + 3 * math.sqrt(3),1],
                        [5,1 + 3 * math.sqrt(3),1],
                        [7,1 + 3 * math.sqrt(3),1]]
```

```python
geometry.material = fresnel.material.Material(color=fresnel.color.linear([0.25,0.5,0.9]),
                                              roughness=0.8)
```

```python
fresnel.preview(scene)
```
