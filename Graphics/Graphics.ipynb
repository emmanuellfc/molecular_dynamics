```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set_style('whitegrid')
```

## Datos experimentales y errores

```python
# Temperature Data

temp = np.array([5.6, 5.7, 5.8, 5.9,
                 6.0, 6.1, 6.2, 6.3,
                 6.4, 6.5, 6.6, 6.7,
                 6.8, 6.9, 7.0]);

# Order Parameters
p2_pure_data = np.array([0.9184768014557397, 0.9049500980820031, 0.9113785945030861, 0.8763553652314207, 
                    0.8848999820728212,  0.8700597175598473, 0.840981959075647, 0.7971109973908581, 
                    0.5083154409773747, 0.39661847887332374, 0.3814405132808837, 0.39612953175369825, 
                    0.40581457347168887,0.38851824973595966, 0.3876542585340579
                        ]);
p2_np_data = np.array([0.9131034204788132, 0.9050319913385602, 0.89217722457266, 0.8856167620106937,
                  0.8339543951857966, 0.8470433819102443, 0.8175989988703383, 0.7211060651247837,
                  0.4592007607826411, 0.39369171725520213, 0.3802709630282621, 0.37546617741613497,
                  0.39825756910661, 0.39428023272121737, 0.3801797185978366]);

# Error data
p2dat_error_pure = 0.1 * np.array([0.04309198284203308, 0.06191793881381956, 0.0310272180094165, 0.0835770969249452, 
                                0.06869670216152579, 0.08186047830225879, 0.08098601311923856, 0.117942399263631, 
                                0.19462185907449284, 0.13230577770433427, 0.1430509901823126, 0.1424952827400811, 
                                0.137421756008749, 0.13226377528495925, 0.13052493131337967]);

p2_error_NP  = 0.1 * np.array([0.0680380251767877, 0.054990657935378946, 0.07462580219893135, 0.061500670017983664,
                               0.10211281722871868, 0.08904967055136948, 0.10738106990483672, 0.1500421199701396,
                               0.1596324498743816, 0.13493566657986214, 0.1415774252509016, 0.12575724573666505,
                               0.13709845856853473, 0.13831217165386023, 0.11729752944720345]);
```

## Interpolación de los datos

```python
t = np.linspace(5.6, 7.0, 500);
p_2_pure = 0.639793 + 0.249493 * np.tanh(12.5837 * (6.35453 - t));
p_2_np = 0.630298 + 0.246851 * np.tanh(10.5989 * (6.32881 - t));
```

## Figuras

## Sistema Puro

```python
# Defining the figure and figure size
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the error bars
ax.errorbar(temp, p2_pure_data, yerr=p2dat_error_pure, label = 'Datos',fmt='o', ecolor='blue',
            color='steelblue', capsize=2)

# Adding a Plot
ax.plot(t, p_2_pure, label='Ajuste', color = 'red');

# Adding plotting parameters
ax.set_title('Sistema Puro', fontsize=14)
ax.set_xlabel('Temperatura', fontsize=12);
ax.set_ylabel('Parámetro de Orden', fontsize=12);
ax.legend();
fig.savefig("sistema_puro.png",dpi = 200, format = "png")
```

### Mezcla binaria

```python
# Defining the figure and figure size
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the error bars
ax.errorbar(temp, p2_np_data, yerr=p2_error_NP, label = 'Datos',fmt='o', ecolor='red',
            color='steelblue', capsize=2)

# Adding a Plot
ax.plot(t, p_2_np, label='Ajuste', color = 'green');

# Adding plotting parameters
ax.set_title('Mezcla binaria', fontsize=16)
ax.set_xlabel('Temperatura', fontsize=13);
ax.set_ylabel('Parámetro de Orden', fontsize=13);
ax.legend();
fig.savefig("mezcla_binaria.png",dpi = 200, format = "png")
```

### Comparación

```python
# Defining the figure and figure size
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the error bars

ax.errorbar(temp, p2_pure_data, yerr=p2dat_error_pure, label = 'Datos Sistema Puro', fmt='o', color = 'C4',ecolor = "blue" , capsize=2)
ax.errorbar(temp, p2_np_data, yerr=p2_error_NP, label = 'Datos Mezcla Binaria', fmt='o', color = 'steelblue', ecolor = 'red', capsize=2)


# Adding a Plot
ax.plot(t, p_2_pure, label='Ajustiste Sistema Puro', color = 'red');
ax.plot(t, p_2_np, label='Ajuste Mezcla Binaria', color = 'green');

# Adding plotting parameters
ax.set_title('Comparación', fontsize=16)
ax.set_xlabel('Temperatura', fontsize=13);
ax.set_ylabel('Parámetro de Orden', fontsize=13);
ax.legend();
fig.savefig("comparación.png",dpi = 200, format = "png")
```
